```python

# Initialize TavilyMap with custom settings
tavily_map = TavilyMap(
    max_depth=5,        # Crawl up to 3 levels deep
    max_breadth=20,     # Follow up to 15 links per page
    max_pages=70        # Limit to 50 total pages for demo
)

print("âœ… TavilyMap initialized successfully!")
```

```python
# Example website to map
demo_url = "https://docs.langchain.com/oss/python/langgraph/overview"

console.print(f"ðŸ” Mapping website structure for: [bold white]{demo_url}[/bold white]", style="bold bright_cyan")
console.print("[bold bright_white]This may take a moment...[/bold bright_white]")

# Map the website structure
site_map = tavily_map.invoke(demo_url)

# Display results
urls = site_map.get('results', [])
console.print(f"\n[bold bright_green]âœ… Successfully mapped [bright_white]{len(urls)}[/bright_white] URLs![/bold bright_green]")

# Show first 50 URLs as examples, using bright colors
console.print("\n[bold bright_yellow]ðŸ“‹ First [bright_white]50[/bright_white] discovered URLs:[/bold bright_yellow]")
for i, url in enumerate(urls[:50], 1):
    console.print(f"  [bright_cyan]{i:2d}[/bright_cyan]. [bright_white][underline]{url}[/underline][/bright_white]")

if len(urls) > 50:
    console.print(f"  [bright_magenta]... and [bright_white]{len(urls) - 50}[/bright_white] more URLs[/bright_magenta]")

```

```python
# Initialize TavilyExtract
tavily_extract = TavilyExtract()

print("âœ… TavilyExtract initialized successfully!")
```

```python

# Select a few interesting URLs for extraction
sample_urls = urls[:2]  # Take the first 2 URLs as a flat list
console.print(f"ðŸ“š Extracting content from [bright_white]{len(sample_urls)}[/bright_white] URLs...", style="bold bright_cyan")

# Extract content
extraction_result = await tavily_extract.ainvoke(input={"urls": sample_urls})

# Display results
extracted_docs = extraction_result.get('results', [])
console.print(f"\n[bold bright_green]âœ… Successfully extracted [bright_white]{len(extracted_docs)}[/bright_white] documents![/bold bright_green]")

# Show summary of each extracted document
for i, doc in enumerate(extracted_docs, 1):
    url = doc.get('url', 'Unknown')
    content = doc.get('raw_content', '')

    # Create a panel for each document with light/bright colors
    panel_content = f"""[bright_white]URL[/bright_white]: [bright_cyan]{url}[/bright_cyan]
[bright_white]Content Length[/bright_white]: [bright_magenta]{len(content):,}[/bright_magenta] characters
[bright_white]Preview[/bright_white]: [bright_yellow]{content}...[/bright_yellow]"""

    console.print(
        Panel(
            panel_content,
            title=f"[bold bright_cyan]Document {i}[/bold bright_cyan]",
            border_style="bright_magenta"
        )
    )
    print()  # Add spacing
```

```python
import asyncio
def chunk_urls(urls: List[str], chunk_size: int = 3) -> List[List[str]]:
    """Split URLs into chunks of specified size."""
    chunks = []
    for i in range(0, len(urls), chunk_size):
        chunk = urls[i:i + chunk_size]
        chunks.append(chunk)
    return chunks

async def extract_batch(urls: List[str], batch_num: int) -> List[Dict[str, Any]]:
    """Extract documents from a batch of URLs."""
    try:
        console.print(f"ðŸ”„ Processing batch {batch_num} with {len(urls)} URLs", style="blue")
        docs = await tavily_extract.ainvoke(input={"urls": urls})
        results = docs.get('results', [])
        console.print(f"âœ… Batch {batch_num} completed - extracted {len(results)} documents", style="green")
        return results
    except Exception as e:
        console.print(f"âŒ Batch {batch_num} failed: {e}", style="red")
        return []

# Process a larger set of URLs in batches
url_batches = chunk_urls(urls[:9], chunk_size=3) # Take first 9 URLs for batch demo, split to batches of 3

console.print(f"ðŸ“¦ Processing 9 URLs in {len(url_batches)} batches", style="bold yellow")

# Process batches concurrently
tasks = [extract_batch(batch, i + 1) for i, batch in enumerate(url_batches)]
batch_results = await asyncio.gather(*tasks)

# Flatten results
all_extracted = []
for batch_result in batch_results:
    all_extracted.extend(batch_result)

console.print(f"\nðŸŽ‰ Batch processing complete! Total documents extracted: {len(all_extracted)}", style="bold green")
```
